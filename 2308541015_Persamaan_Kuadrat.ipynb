```python
#TUGAS ALGORITMA DAN STRUKTUR DATA
#akan diselesaikan 2 kasus persamaan kuadrat sebagai berikut:
#menentukan persamaan kuadrat jika diketahui nilai a,b,c => kasus 1 pada kode
#kemana arah terbuka parabola (keatas atau kebawah)
#bagaimana kondisi parabola memotong sumbu y (potong sumbu y positif, potong sumbu y di pusat (0,0), potong sumbu y negatif)
#macam-macam akar persamaan kuadrat dilihat dari diskriminan
#penyelesaian persamaan kuadrat saat menggunakan rumus ABC
#menentukan titik puncak dan sumbu simetri
#menentukan posisi titik puncak (dikiri sumbu y, pada sumbu y, dikanan sumbu y)
#menentukan persamaan kuadrat jika diketahui akar akarnya => kasus 2 pada kode
#menyusun persamaan kuadrat baru jika diketahui akan-akarnya
#menentukan grafik fungsi kuadrat
#arsir daerah himpunan penyelesaiannya
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n ")
print("Ayo buat persamaan kuadrat dengan 2 kasus dibawah ini! ")
kasus_15 = float(input("pilih antara kasus 1 atau kasus 2! \n"))

if kasus_15 == 1 : #menentukan persamaan kuadrat jika diketahui nilai a,b,c
  print("membuat persamaan kuadrat dengan input nilai a,b dan c")
  print("ayo kita input nilai-nilai pembentuk persamaan kuadratnya! ")
  print("syarat nilai a tidak boleh 0 yaa :) \n")
  a_15 = float(input("masukkan koefisien x kuadrat! ")) #syarat nilai a tidak boleh 0
  b_15 = float(input("masukkan koefisien x! "))
  c_15 = float(input("masukkan konstanta! "))
  print("\n")
  #kemungkinan yang terjadi akibat dari nilai a
  if a_15 == 0:
    print("nilai a tidak boleh 0 yaa, karena nanti tidak jadi persamaan kuadrat :) \n")
  elif a_15 > 0:
    print("wahh kurvanya akan membentuk keatas nihh")
  else:
    print("wahh kurvanya akan membentuk kebawah nihh")
  #kemungkinan yang terjadi akibat dari nilai c
  if c_15 == 0:
    print("parabola memotong sumbu y di titik asal (0,0)")
    print("ayoo kita gambar persmaannya \n")
  elif c_15 > 0:
    print("parabola memotong sumbu y pada nilai positif")
    print("ayoo kita gambar persmaannya \n")
  else:
    print("parabola memotong sumbu y pada nilai negatif")
    print("ayoo kita gambar persmaannya \n")
  print(f"persamaan kuadratnya adalah: y = {a_15}x^2 + {b_15}x + {c_15} \n")

  #sekarang hitung akar-akar persamaannya
  D_15 = (b_15**2) - (4*a_15*c_15) #hitung terlebih dahulu diskriminannya
  print(f"diskriminan dari persamaan diatas adalah: {D_15} \n")
  if D_15 == 0:
    print("solusi akar-akar persamaan trivial atau akar-akarnya kembar :) \n")
    #dengan menggunakan rumus ABC, maka akar-akar penyelesaiannya
    akar1_15 = ((-b_15) + D_15**(1/2)) / 2*a_15
    akar2_15 = ((-b_15) - D_15**(1/2)) / 2*a_15
    print(f"akar-akar persamaannya adalah: {akar1_15} dan {akar2_15}" )
    #hitung sumbu simetrisnya
    xs_15 = -b_15 /2*a_15
    print (f"sumbu simetrisnya adalah: {xs_15}")
    #kemungkinan yang terjadi akibat nilai sumbu simetris
    if xs_15 < 0:
        print("Titik Puncak di Kiri Sumbu Y")
    elif xs_15 == 0:
        print("Titik Puncak di Sumbu Y")
    else:
        print("Titik Puncak di Kanan Sumbu Y")
    #hitung titik puncaknnya
    yp_15 = a_15*xs_15**2 + b_15*xs_15 + c_15
    print (f"titik puncaknya adalah: ({xs_15},{yp_15}) \n")
    #gambar grafik persamaannya
    x_15 = np.linspace(xs_15-5, xs_15+5, 400)
    y_15 = a_15*x_15**2 + b_15*x_15 + c_15
    plt.plot(x_15, y_15, label=f"{a_15}x^2 + {b_15}x + {c_15} = 0", color="pink")
    plt.axhline(0, color="black")
    plt.axvline(0, color="black")
    plt.scatter([akar1_15, akar2_15], [0, 0], color="red")
    plt.text(akar1_15, 0, f"({akar1_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
    plt.text(akar2_15, 0, f"({akar2_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
    plt.axvline(x=xs_15, color="plum", linestyle="--", label=f"Sumbu Simetri: x={xs_15:.2f}")
    plt.scatter(xs_15, yp_15, color="darkred")
    plt.text(xs_15, yp_15, f"({xs_15}, {yp_15})", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
    print("Ayo arsir daerah himpunan penyelesaiannya! ")
    print("Ada 2 pilihan nihh \n 1. Daerah himpunan penyelesaian saat y<0 \n 2. Daerah himpunan penyelesaian saat y>0 \n")
    hp_15 = float(input("pilih antara HP 1 atau HP 2! \n"))
    if hp_15 == 1:
      print("ayo arsir daerah himpunan peenyelesaian saat y<0")
      plt.fill_between(x_15, y_15, where=(y_15 <= 0), color="orchid", alpha=0.3)
    elif hp_15 == 2:
      print("ayo arsir daerah himpunan peenyelesaian saat y<0")
      plt.fill_between(x_15, y_15, where=(y_15 >= 0), color="darkorchid", alpha=0.3)
    else:
      print("ingat hanya ada 2 pilihan yaa :) ")
    plt.title("Grafik Persamaan Kuadrat")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
  else:
    print("wahh ada 2 kemungkinan, yaitu akar-akar imajiner dan terdapat 2 solusi akar \n")
    if D_15 > 0:
      print("karena diskriminannya bernilai positif")
      print("ayo kita hitung akar-akar persamaannya! \n")
      #dengan menggunakan rumus ABC, maka akar-akar penyelesaiannya
      akar1_15 = ((-b_15) + D_15**(1/2)) / 2*a_15
      akar2_15 = ((-b_15) - D_15**(1/2)) / 2*a_15
      print(f"akar-akar persamaannya adalah: {akar1_15} dan {akar2_15}")
      #hitung sumbu simetrisnya
      xs_15 = -b_15 /2*a_15
      print (f"sumbu simetrisnya adalah: {xs_15}")
      #kemungkinan yang terjadi akibat nilai sumbu simetris
      if xs_15 < 0:
          print("Titik Puncak di Kiri Sumbu Y")
      elif xs_15 == 0:
          print("Titik Puncak di Sumbu Y")
      else:
          print("Titik Puncak di Kanan Sumbu Y")
      #hitung titik puncaknnya
      yp_15 = a_15*xs_15**2 + b_15*xs_15 + c_15
      print (f"titik puncaknya adalah: ({xs_15},{yp_15}) \n")
      #gambar grafik persamaannya
      x_15 = np.linspace(xs_15-5, xs_15+5, 400)
      y_15 = a_15*x_15**2 + b_15*x_15 + c_15
      plt.plot(x_15, y_15, label=f"{a_15}x^2 + {b_15}x + {c_15} = 0", color="pink")
      plt.axhline(0, color="black")
      plt.axvline(0, color="black")
      plt.scatter([akar1_15, akar2_15], [0, 0], color="red")
      plt.text(akar1_15, 0, f"({akar1_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
      plt.text(akar2_15, 0, f"({akar2_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
      plt.axvline(x=xs_15, color="plum", linestyle="--", label=f"Sumbu Simetri: x={xs_15:.2f}")
      plt.scatter(xs_15, yp_15, color="darkred")
      plt.text(xs_15, yp_15, f"({xs_15}, {yp_15})", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
      print("Ayo arsir daerah himpunan penyelesaiannya! ")
      print("Ada 2 pilihan nihh \n 1. Daerah himpunan penyelesaian saat y<0 \n 2. Daerah himpunan penyelesaian saat y>0 \n")
      hp_15 = float(input("pilih antara HP 1 atau HP 2! \n"))
      if hp_15 == 1:
        print("ayo arsir daerah himpunan peenyelesaian saat y<0")
        plt.fill_between(x_15, y_15, where=(y_15 <= 0), color="orchid", alpha=0.3)
      elif hp_15 == 2:
        print("ayo arsir daerah himpunan peenyelesaian saat y<0")
        plt.fill_between(x_15, y_15, where=(y_15 >= 0), color="darkorchid", alpha=0.3)
      else:
        print("ingat hanya ada 2 pilihan yaa :) ")
      plt.title("Grafik Persamaan Kuadrat")
      plt.xlabel("x")
      plt.ylabel("y")
      plt.legend()
      plt.grid(True)
      plt.show()
    else:
      print("karena diskriminannya bernilai negatif")
      print("akar-akar imajiner, tidak terdapat akar-akar penyelesaian dehh \n")

elif kasus_15 == 2 : #menentukan persamaan kuadrat jika diketahui akar akarnya
  print("membuat persamaan kuadrat dengan input akar-akar penyelesaiannya")
  print("ayo kita input akar-akar penyelesaiannya! \n")
  akar1_15 = float(input("masukkan akar penyelesaian pertama! "))
  akar2_15 = float(input("masukkan akar penyelesaian kedua! "))
  #menentukan koefisien persamaan kuadrat ax^2 + bx + c = 0
  a_15 = 1
  b_15 = -(akar1_15 + akar2_15)
  c_15 = akar1_15 * akar2_15
  print(f"persamaan kuadratnya adalah: y = {a_15}x^2 + {b_15}x + {c_15}")
  #hitung sumbu simetrisnya
  xs_15 = -b_15 /2*a_15
  print (f"sumbu simetrisnya adalah: {xs_15}")
  #kemungkinan yang terjadi akibat nilai sumbu simetris
  if xs_15 < 0:
      print("Titik Puncak di Kiri Sumbu Y")
  elif xs_15 == 0:
      print("Titik Puncak di Sumbu Y")
  else:
      print("Titik Puncak di Kanan Sumbu Y")
  #hitung titik puncaknnya
  yp_15 = a_15*xs_15**2 + b_15*xs_15 + c_15
  print (f"titik puncaknya adalah: ({xs_15},{yp_15}) \n")
  #gambar grafik persamaannya
  x_15 = np.linspace(akar1_15-5, akar2_15+5, 400)
  y_15 = a_15*x_15**2 + b_15*x_15 + c_15
  plt.plot(x_15, y_15, label=f"{a_15}x^2 + {b_15}x + {c_15} = 0", color="pink")
  plt.axhline(0, color="black")
  plt.axvline(0, color="black")
  plt.scatter([akar1_15, akar2_15], [0, 0], color="red")
  plt.text(akar1_15, 0, f"({akar1_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.text(akar2_15, 0, f"({akar2_15}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.axvline(x=xs_15, color="plum", linestyle="--", label=f"Sumbu Simetri: x={xs_15:.2f}")
  plt.scatter(xs_15, yp_15, color="darkred")
  plt.text(xs_15, yp_15, f"({xs_15}, {yp_15})", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  print("Ayo arsir daerah himpunan penyelesaiannya! ")
  print("Ada 2 pilihan nihh \n 1. Daerah himpunan penyelesaian saat y<0 \n 2. Daerah himpunan penyelesaian saat y>0 \n")
  hp_15 = float(input("pilih antara HP 1 atau HP 2! \n"))
  if hp_15 == 1:
    print("ayo arsir daerah himpunan peenyelesaian saat y<0")
    plt.fill_between(x_15, y_15, where=(y_15 <= 0), color="orchid", alpha=0.3)
  elif hp_15 == 2:
    print("ayo arsir daerah himpunan peenyelesaian saat y<0")
    plt.fill_between(x_15, y_15, where=(y_15 >= 0), color="darkorchid", alpha=0.3)
  else:
    print("ingat hanya ada 2 pilihan yaa :) ")
  plt.title("Grafik Persamaan Kuadrat dari Akar-Akar Penyelesaian")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.grid(True)
  plt.show()

else: #kemungkinan saat input selain 1 dan 2
  print("ingattt, disini hanya terdapat 2 kasus yaa. Mohon masukkan 1 atau 2 sajaa :) ")
```
