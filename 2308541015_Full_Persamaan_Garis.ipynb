```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", lifestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```

```python
print("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :) \n")
print("Ayo kita selidiki kasus-kasus diatas :) ")

import numpy as np
import matplotlib.pyplot as plt

kasus_15 = float(input("Kasus berapakah yang akan dikerjakan? \n"))

#penyelesaian kasus 1
if kasus_15 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui. \n")
    #masukkan titik-titik garis pertama
    d_15 = float(input("Nilai x1 dari titik pertama dari garis pertama adalah \t: "))
    i_15 = float(input("Nilai y1 dari titik pertama dari garis pertama adalah \t: "))
    a_15 = float(input("Nilai x2 dari titik kedua dari garis pertama adalah \t: "))
    n_15 = float(input("Nilai y2 dari titik kedua dari garis pertama adalah \t: "))
    #masukkan titik-titik garis kedua
    b_15 = float(input("Nilai x1 dari titik pertama dari garis kedua adalah \t: "))
    e_15 = float(input("Nilai y1 dari titik pertama dari garis kedua adalah \t: "))
    s_15 = float(input("Nilai x2 dari titik kedua dari garis kedua adalah \t: "))
    t_15 = float(input("Nilai y2 dari titik kedua dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = (n_15-i_15)/(a_15-d_15) # Gradien garis pertama
    m2_15= (t_15-e_15)/(s_15-b_15) # Gradien garis kedua
    #mencari titik potong garis terhadap sumbu y
    tp1_15 = -m1_15*d_15+i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = -m2_15*b_15+e_15 # Titik potong garis kedua terhadap sumbu y

    print("\n")
    #persamaan garis pertama adalah y=m1_15(x-d_15)+i_15=m1_15x-m1_15(d_15+i_15)
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    #persamaan garis kedua adalah y=m2_15(x-b_15)+e_15=m2_15x-m2_15(b_15+e_15)
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
      print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
      #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
      if tp1_15 == tp2_15 :
        print("wahh garisnya berhimpit") #garisnya berhimpit
        print("maka memiliki banyak solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berhimpit")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
      else:
        print("wahh garisnya sejajar") #garisnya sejajar
        print("maka tidak memiliki solusi penyelesaian deh \n")
        x_15 = np.linspace(-5,5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Sejajar")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
        print("ayo kita gambar dan cari titik potongnyaa \n")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x = 0,color='black')
        plt.axhline(y = 0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 2
elif kasus_15 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    #gradien dan titik-titi persamaaan garis 1
    m1_15 = float(input("Masukkan Gradien dari garis pertama\t: "))
    d_15 = float(input("Nilai x dari titik yang dilewati oleh garis pertama\t: "))
    i_15 = float(input("Nilai y dari titik yang dilewati oleh garis pertama\t: "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_15,d_15,i_15))
    #gradien dan titik-titi persamaaan garis 1
    m2_15 = float(input("Masukkan Gradien dari garis kedua \t: "))
    b_15 = float(input("Nilai x dari titik yang dilewati oleh garis kedua \t: "))
    e_15 = float(input("Nilai y dari titik yang dilewati oleh garis kedua \t: "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_15,b_15,e_15))

    #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
    tp1_15 = -m1_15*d_15+i_15
    tp2_15 = -m2_15*b_15+e_15

    print("\n")
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_15,tp1_15))
    print("Persamaan garis kedua adalah y={}x+{}\n".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit") #garisnya berhimpit
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar") #garisnya sejajar
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="darkblue", linestyle = "--", label = "y={}x+{}".format(m1_15,tp1_15))
            plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="skyblue", label = "y={}x+{}".format(m2_15,tp2_15))
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
        tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
        tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
        print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")
        x_15 = np.linspace(tp_x_15-5,tp_x_15+5,100)
        plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color='pink',label="y={}(x-{})+{}".format(m1_15,d_15,i_15))
        plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color='skyblue',label="y={}(x-{})+{}".format(m2_15,b_15,e_15))
        plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4)
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

#penyelesaian kasus 3
elif kasus_15 == 3:
    print("Kedua garis memenuhi persamaan berbentuk dx+iy=a dan bx+ey=s.")
    #masukkan elemen-elemen pembuat persamaan kedua
    d_15 = float(input("Koefisien dari x pada garis pertama adalah \t: "))
    i_15 = float(input("Koefisien dari y pada garis pertama adalah \t: "))
    a_15 = float(input("Konstanta dari garis pertama adalah \t: "))
    #masukkan elemen-elemen pembuat persamaan kedua
    b_15 = float(input("Koefisien dari x pada garis kedua adalah \t: "))
    e_15 = float(input("Koefisien dari y pada garis kedua adalah \t: "))
    s_15 = float(input("Konstanta dari garis kedua adalah \t: "))
    #mencari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_15 = ((-d_15)/i_15) # Gradien garis pertama
    m2_15= (-b_15)/(e_15) # Gradien garis kedua

    #mencari titik potong garis terhadap sumbu y
    tp1_15 = a_15/i_15 # Titik potong garis pertama terhadap sumbu y
    tp2_15 = s_15/e_15 # titik potong garis kedua terhadap sumbu y garis kedua

    print("persamaan garis 1 adalah \t: y= ({}x)+{}".format(m1_15,tp1_15))
    print("persamaan garis 2 adalah \t: y= ({}x)+{}".format(m2_15,tp2_15))

    if m1_15 == m2_15:
        print("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
        if tp1_15 == tp2_15:
            print("wahh garisnya berhimpit")
            print("maka memiliki banyak solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("wahh garisnya sejajar")
            print("maka tidak memiliki solusi penyelesaian deh")
            x_15 = np.linspace(-5,5,100)
            plt.plot(x_15,(m1_15*x_15)+tp1_15,color="darkblue", linestyle = "--", label="line 1")
            plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="line 2")
            plt.axvline(x=0,color='black')
            plt.axhline(y=0,color='black')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan di satu titik dehh")
        print("ayo kita hitung titik potongnya dengan matriks! ")
        #tentukan matriks koefisien A dan matriks konstanta B
        A_15 = np.array([[d_15, b_15], [i_15,e_15]])
        B_15 = np.array([a_15, s_15])
        print("Matriks dari koefisien x dan y yaitu: \n", A_15)
        print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_15)
        #untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
        #akan dihitung invers dari matriks A dengan rumus numpy
        Ainv_15 = np.linalg.inv(A_15)
        print("Invers dari Matriks A yaitu: \n", Ainv_15)
        # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
        xy_15 = np.dot((Ainv_15),(B_15))
        print("titik potongnya adalah : \n [x,y]=", xy_15)
        #ayo gambar persamaaan garisnya
        x_15 = np.linspace(xy_15-5,xy_15+5,100)
        plt.plot(x_15,(m1_15*x_15)+tp1_15,color="pink",label="y= ({}x)+{}".format(m1_15,tp1_15))
        plt.plot(x_15,(m2_15*x_15)+tp2_15,color="skyblue",label="y= ({}x)+{}".format(m2_15,tp2_15))
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("ingattt, disini hanya terdapat 3 kasus yaa. Mohon masukkan 1, 2 atau 3 sajaa :) ")
```
