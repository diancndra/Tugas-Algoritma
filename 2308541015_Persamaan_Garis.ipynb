```python
print ("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :)")
print ("AYO GAMBAR PERSAMAAN GARIS!")

import numpy as np
import matplotlib.pyplot as plt
#persamaan garis ini melewati titik (e_15,f_15) dan (g_15,h_15)
e_15 = -1 #titik (e_15,f_15)
f_15 = 2
g_15 = -4 #titik (g_15,h_15)
h_15 = 5

#setelah titik-titiknya diketahui, kemudian hitung gradiennya
gradient_15 =(h_15-f_15)/(g_15-e_15)
print ("gradien garis 1 adalah: y= {}".format(gradient_15))

#setelah mendapatkan gradien, subtitusi gradien dan titiknya ke persamaan y_15=m_15*(x_15-e_15)+f_15 dengan m_15 = gradient
x_15 = np.arange(0,5,0.01) #tentukan batas koordinatnya
plt.figure(figsize=(4,4)) #tentukan besar gambar garisnya
plt.plot(x_15,gradient_15*(x_15-e_15)+f_15) #definisikan persamaan garisnya
print ("persamaan garis 1 adalah: y= {}x+{}".format(gradient_15,(-gradient_15*e_15)+f_15))
plt.plot(x_15,gradient_15*(x_15-e_15)+f_15,color="green",label="y= {}x+{}".format(gradient_15,(-gradient_15*e_15)+f_15))
plt.legend(loc= "upper right") #posisikan dikiri atas agar terlihat rapi
plt.title("Gambar Persamaan Garis") #beri judul gambar persamaan garisnya
plt.grid(True) #tambahkan garis bantu
plt.xlabel("x") #berikan label sumbu x
plt.ylabel("y") #berikan label sumbu y
plt.show()
```

```python
print ("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :)")
print ("Apabila terdapat 2 persamaan garis, apa yang akan terjadi??")
print ("Ayo kita cari tauu! \n")

import numpy as np
import matplotlib.pyplot as plt
#persamaan garis 1 melewati titik (d_15,i_15) dan (a_15,n_15)
d_15 = 2 #titik (d_15,i_15)
i_15 = 0
a_15 = 3 #titik (a_15,n_15)
n_15 = 4
#hitung gradien persamaan garis 1
m1_15 =(n_15-i_15)/(a_15-d_15)
print ("jika gradien garis 1 adalah: m1 = {}".format(m1_15))
#subtitusi gradien dan titik titik yang diketahui ke y1_15=m1_15*(x_15-d_15)+i_15
print ("maka persamaan garis 1 adalah: y1 = {}x+{}".format(m1_15,-m1_15*d_15+i_15)+"\n")

#persamaan garis 2 melewati titik (b_15,e_15) dan (s_15,t_15)
b_15 = 2 #titik(b_15,e_15)
e_15 = 3
s_15 = 4 #titik(s_15,t_15)
t_15 = 7
#hitung gradien persamaan garis 2
m2_15 =(t_15-e_15)/(s_15-b_15)
print ("jika gradien garis 2 adalah: m2 = {}".format(m2_15))
#subtitusi gradien dan titik titik yang diketahui ke y2_15=m2_15*(x_15-b_15)+e_15
print ("maka persamaan garis 2 adalah: y2 = {}x+{}".format(m2_15,-m2_15*b_15+e_15)+"\n")

#kemungkinan yang terjadi apabila terdapat 2 persamaan garis
if m1_15 == m2_15 :
  print ("waahh ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit nihh")
else :
  print ("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
  print ("ayo kita gambar dan cari titik potongnyaa \n")
  #mencari titik potong persamaan garis (tp_x,tp_y)
  tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
  tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
  print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")

#gambar persamaan garis beserta titik potongnya
x_15 = np.arange(-2,8,0.01) #tentukan batas koordinat sumbu x
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="skyblue",label="y1 = {}x+{}".format(m1_15,-m1_15*d_15+i_15))
plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="pink",label="y2 = {}x+{}".format(m2_15,-m2_15*b_15+e_15))
plt.plot(tp_x_15,tp_y_15,color="black",marker="o",markersize=4) #tandai titik potongnya
plt.title("Gambar Titik Potong Persamaan Garis") #berikan judul gambarnya
plt.legend(loc="best") #posisikan ditempat terbaik
plt.grid(True) #tambahkan garis bantu
plt.xlabel("x") #berikan label sumbu x
plt.ylabel("y") #berikan label sumbu y
plt.show()
```

```python
print ("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :)")
print ("Gimana jadinya kalau persamaan garis sebelumnya kita arsir yaa??")
print ("Ayo kita arsir bersamaa!\n")

import numpy as np
import matplotlib.pyplot as plt
x_15 = np.linspace(-2, 8, 1000)
#definisikan persamaan garis sebelumnya
def line1(x_15):
  return 4*x_15-8 #persamaan garis 1
print ("persamaan garis pertama adalah \t: y1=4x-8")
def line2(x_15):
  return 2*x_15-1 #persamaan garis 2
print ("persamaan garis kedua adalah \t: y2=2x-1")
fig, ax = plt.subplots() #gunakan fungsi figure dan axes untuk menggambar
#beri label dan warna kedua persamaan garis
ax.plot(x_15, line1(x_15), label='Line 1',color="skyblue")
ax.plot(x_15, line2(x_15), label='Line 2',color="pink")
#definisikan y1 dan y2 dengan definisi persamaan garis diatas
y1_15 = line1(x_15)
y2_15 = line2(x_15)

#arsir daerah diantara kedua garis
print ("ayo kita arsir daerah diantara kedua garis diatas! \n ")
#daerah yang diarsir saat persamaan garis 1 lebih besar dari persamaan garis 2
ax.fill_between(x_15, y1_15, y2_15, where=(y1_15 > y2_15), color='pink',alpha=1)
#daerah yang diarsir saat persamaan garis 1 lebih kecil dari persamaan garis 2
ax.fill_between(x_15, y1_15, y2_15, where=(y1_15 < y2_15), color='skyblue',alpha=1)
ax.set_title('Daerah Yang Diarsir Dari Persamaan Garis') #berikan judul gambar
ax.legend() #posisinya menyesuaikan
plt.grid(True) #tambahkan garis bantu
plt.show()
```

```python
print ("SELAMAT DATANG DI PROGRAM SEDERHANA DIAN :)")
print ("Apabila terdapat 2 persamaan garis, apa yang akan terjadi??")
print ("Ayo kita cari tauu! \n")

import numpy as np
import matplotlib.pyplot as plt
#persamaan garis 1 melewati titik (d_15,i_15) dan (a_15,n_15)
d_15 = 1 #titik (d_15,i_15)
i_15 = 1
a_15 = 3 #titik (a_15,n_15)
n_15 = 3
#hitung gradien persamaan garis 1
m1_15 =(n_15-i_15)/(a_15-d_15)
print ("jika gradien garis 1 adalah: m1 = {}".format(m1_15))
#subtitusi gradien dan titik titik yang diketahui ke y1_15=m1_15*(x_15-d_15)+i_15
y1_15 = (m1_15,-m1_15*d_15+i_15)
print ("maka persamaan garis 1 adalah: y1 = {}x+{}".format(m1_15,-m1_15*d_15+i_15)+"\n")

#persamaan garis 2 melewati titik (b_15,e_15) dan (s_15,t_15)
b_15 = 2 #titik(b_15,e_15)
e_15 = 2
s_15 = 6 #titik(s_15,t_15)
t_15 = 6
#hitung gradien persamaan garis 2
m2_15 =(t_15-e_15)/(s_15-b_15)
print ("jika gradien garis 2 adalah: m2 = {}".format(m2_15))
#subtitusi gradien dan titik titik yang diketahui ke y2_15=m2_15*(x_15-b_15)+e_15
y2_15 = (m2_15,-m2_15*b_15+e_15)
print ("maka persamaan garis 2 adalah: y2 = {}x+{}".format(m2_15,-m2_15*b_15+e_15)+"\n")

#kemungkinan yang terjadi apabila terdapat 2 persamaan garis
if m1_15 == m2_15 :
  print ("ternyata gradien keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit")
  #saat m1_15=m2_15 maka terdapat 2 kemungkinan lagi, yaitu saat persamaannya sama dan persamaannya tidak sama
  if y1_15 == y2_15 : #garisnya berhimpit
    print ("wahh garisnya berhimpit")
    print ("maka memiliki banyak solusi penyelesaian deh")
  else : #garisnya sejajar
    print ("wahh garisnya sejajar") #garisnya sejajar
    print ("maka tidak memiliki solusi penyelesaian deh")
else : #garisnya sudah pasti berpotongan
  print ("nahh gradiennya tidak samaa, jadi sudah pasti garisnya berpotongan")
  print ("ayo kita gambar dan cari titik potongnyaa \n")
  #mencari titik potong persamaan garis (tp_x_15,tp_y_15)
  tp_x_15 = (m1_15*d_15-i_15-m2_15*b_15+e_15)/(m1_15-m2_15) #titik potong persamaan garis 1
  tp_y_15 = m2_15*(tp_x_15-b_15)+e_15 #titik potong persamaan garis 2
  print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_15,tp_y_15)+"\n")

#gambar persamaan garis
x_15 = np.arange(-2,8,0.01)
plt.figure(figsize=(6,4),dpi=100)
if -m1_15*d_15+i_15>0 :
  plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="lightblue",label="y1={:.2f}x+{:.2f}".format(m1_15,-m1_15*d_15+i_15))
else:
  plt.plot(x_15,m1_15*(x_15-d_15)+i_15,color="yellow",label="y1={:.2f}x{:.2f}".format(m1_15,-m1_15*d_15+i_15))
if -m2_15*b_15+e_15>0:
  plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="navy",linestyle="dashed", label="y2={:.2f}x+{:.2f}".format(m2_15,-m2_15*b_15+e_15))
else:
  plt.plot(x_15,m2_15*(x_15-b_15)+e_15,color="blue",linestyle="dashed", label="y2={:.2f}x{:.2f}".format(m2_15,-m2_15*b_15+e_15))

plt.title("Gambar Persamaan Garis Berhimpit")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
