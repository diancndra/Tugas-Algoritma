```python
#TUGAS ALGORITMA
#Jika diketahui bahwa ğ‘› adalah dua digit terakhir dari NIM kalian
#Maka tentukan bilangan prima antara 1000(ğ‘›)Â danÂ 1000(ğ‘›+1)
#Jika NIM = 2308541015, maka n = 15 dan n+1 = 16

print("Jika n=15, maka akan dicari bilangan prima antara 1000ğ‘›Â danÂ 1000(ğ‘›+1)")

#Definisikan bilangan prima
def bilangan_prima_15(x_15):
  if x_15 <= 1:
    return False
  for i_15 in range(2, int(x_15**0.5) + 1):
    if x_15 % i_15 == 0:
      return False
  else:
    return True

#Masukkan nilai n sesuai arahan pada soal
n_15 = 15

#Gunakan fungsi primes menyimpan semua bilangan prima yang ditemukan dalam rentang yang ditentukan
primes_15 = []

#Tentukan rentang bilangan prima yang akan ditemukan yaitu antara 1000(ğ‘›)Â danÂ 1000(ğ‘›+1)
rentang_bawah_15 = 1000 * n_15
rentang_atas_15 = 1000 * (n_15 + 1)
print("Rentang bilangan prima yang akan dicari adalah {} dan {}\n".format(rentang_bawah_15,rentang_atas_15))

for angka_15 in range(rentang_bawah_15, rentang_atas_15):
    if bilangan_prima_15 (angka_15):
       #Gunakan append untuk menambahkan setiap bilangan prima yang ditemukan ke dalam rentang
      primes_15.append(angka_15)

#Print untuk mendapatkan output dari kode diatas
print("Bilangan prima antara 15000 dan 16000 adalah:\n", primes_15)
#Gunakan fungsi len untuk mengetahui banyak bilangan prima pada rentang 15000 - 16000
print("Jumlah bilangan prima dalam rentang tersebut adalah:", len(primes_15))
```

```python
#Karena jumlah biangan primanya cukup banyak, maka outputnya akan dibagi menjadi 4 kolom
print("Bilangan prima antara 15000 dan 16000 adalah:\n")
total_primes_15 = len(primes_15) #jumlah bilangan bilangan prima pada rentang ynag ditentukan
seperempat_15 = (total_primes_15) // 4 #karena akan dibuat menjadi 4 kolom maka bagi jumlah bilangan prima dengan 4
for i_15 in range(seperempat_15):
    if i_15 < total_primes_15:
        print(f"{primes_15[i_15]}", end=' ') #kolom pertama
    if i_15 + seperempat_15 < total_primes_15:
        print(f"\t{primes_15[i_15 + seperempat_15]}", end=' ') #kolom kedua
    if i_15 + 2 * seperempat_15 < total_primes_15:
        print(f"\t{primes_15[i_15 + 2 * seperempat_15]}", end=' ') #kolom ketiga
    if i_15 + 3 * seperempat_15 < total_primes_15:
        print(f"\t{primes_15[i_15 + 3 * seperempat_15]}") #kolom keempat

print("\nJumlah bilangan prima dalam rentang tersebut adalah:", total_primes_15)
```
