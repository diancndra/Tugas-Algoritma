```python
berat = float(input("berapa berat anda? "))
kg_atau_ons = input("beratnya dalam kg atau ons? ")

if kg_atau_ons.upper() == "KG" :
  berat = berat*35.274
  print ("Berat anda adalah " + str(berat) + " ons")
elif kg_atau_ons.upper() == "ONS" :
  berat = berat/35.274
  print ("Berat anda adalah " + str(berat) + " kg")
```
