```python
#MENCARI NILAI N DENGAN NUMPY
import numpy as np
print("ayo cari nilai n dari persamaan berikut!")
print("x^2024-x^2025 = (x(1-x))^n")
print("dengan nilai x1 = 0.6180339887498949 \ndan nilai x2 = -1.618033988749895\n")

#mencari nilai n dengan substitusi nilai x1
print("mencari nilai n dengan substitusi nilai x1")
x1_15 = 0.6180339887498949
#ubah bentuk log(x^2024-x^2025) dapat diubah menjadi 2024*log(x)+log(1-x) dengan sifat logaritma
a_15 = 2024*np.log(abs(x1_15)) + np.log(abs(1-x1_15))
#ubah bentuk log(x(1-x)) menjadi log(x)+log(1-x) dengan sifat logaritma
b_15 = np.log(abs(x1_15)) + np.log(abs(1-(x1_15)))
#nilai n dengan logaritma, maka n=log(x^2024-x^2025)/log(x(1-x))
n1_15 = (a_15/b_15)
print("dengan menggunakan sifat logaritma, maka ")
print("nilai n adalah {}".format(n1_15))

#mencari nilai n dengan substitusi nilai x2
print("\nmencari nilai n dengan substitusi nilai x2")
x2_15 = -1.618033988749895
#ubah bentuk log(x^2024-x^2025) dapat diubah menjadi 2024*log(x)+log(1-x) dengan sifat logaritma
d_15 = 2024*np.log(abs(x2_15)) + np.log(abs(1-(x2_15)))
#ubah bentuk log(x(1-x)) menjadi log(x)+log(1-x) dengan sifat logaritma
e_15 = np.log(abs(x2_15)) + np.log(abs(1-(x2_15)))
#nilai n dengan logaritma, maka n=log(x^2024-x^2025)/log(x(1-x))
n2_15 = (d_15/e_15)
print("dengan menggunakan sifat logaritma, maka ")
print("nilai n adalah {}".format(n2_15))
```

```python
#MENCARI NILAI N DENGAN MATH
print("ayo cari nilai n dari persamaan berikut!")
print("x^2024-x^2025 = (x(1-x))^n")
print("dengan nilai x1 = 0.6180339887498949 \ndan nilai x2 = -1.618033988749895\n")

import math
x1_15 = 0.6180339887498949
x2_15 = -1.618033988749895

#akan dicari nilai n dengan logaritma, maka n=log(x^2024-x^2025)/log(x(1-x))

#jabarkan bentuk pembilangnya agar output kode tidak nmerical out of range
#bentuk log(x^2024-x^2025) dapat diubah menjadi 2024*log(x)+log(1-x) dengan sifat logaritma
pembilang1_15 = 2024*math.log(abs(x1_15)) + math.log(abs(1-x1_15))
pembilang2_15 = 2024*math.log(abs(x2_15)) + math.log(abs(1-x2_15))

#jabarkan juga bentuk penyebutnya
#ubah bentuk log(x(1-x)) menjadi log(x)+log(1-x) dengan sifat logaritma
penyebut1_15 = math.log(abs(x1_15)) + math.log(abs(1-(x1_15)))
penyebut2_15 = math.log(abs(x2_15)) + math.log(abs(1-(x2_15)))

#akan dihitung nilai n dari x1 dan x2
n1_15 = pembilang1_15/penyebut1_15
n2_15 = pembilang2_15/penyebut2_15

#print nilai n1 dan n2 setelah disubstitusi nilai x1 dan x2
print("mencari nilai n dengan substitusi nilai x1")
print("dengan menggunakan sifat logaritma, maka ")
print("nilai n saat x1 adalah {}".format(n1_15))

print("\nmencari nilai n dengan substitusi nilai x2")
print("dengan menggunakan sifat logaritma, maka ")
print("nilai n saat x2 adalah {}".format(n2_15))
```
